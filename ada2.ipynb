# General Info

## DFS
procedure DFS(G, v) is
    label v as discovered
    for all directed edges from v to w that are in G.adjacentEdges(v) do
        if vertex w is not labeled as discovered then
            recursively call DFS(G, w)


## General AI model. Some models may differ.

2. Connections graph. Stores memories as images, words, etc.
3. Decision maker. Decides what to do with the results from 1 and 2. Can go back to 1 or 2.
4. Output. Decided by previous step. Only seperate because 3 can go back to 1 and 2.

# Model

```python
class Node:
    def __init__(self, value, connection):
        self.value = value
        self.connections = [connection]
        self.flag = False

class Memory:
    def __init__(self):
        self.memory = []
        self.counter = 0

    def addNode(self, value):
        if self.counter != 0:
            self.memory.append(Node(value, self.memory[self.counter-1]))
        else:
            self.memory.append(Node(value, None))
        self.counter += 1

    def printMemoryValues(self):
        for node in self.memory:
            print(node.value)
    
    def printDfs(self):
        self.dfs(self.memory[0])
                    
    def dfs(self, node):
        node.flag = True
        print(node.value)
        for conn in node.connections:
                if conn != None and conn.flag == True:
                    self.dfs(conn)
                
```
