```python
class Node:
    def __init__(self, name, value):
        self.name = name
        self.value = value
```

```python
class Memory:
    def __init__(self):
        self.memory = []
        self.counter = 0

    def addNode(self, value):
        self.memory.append(Node(self.counter, value))
        self.counter += 1

    def printMemoryValues(self):
        for node in self.memory:
            print(node.value)
```
